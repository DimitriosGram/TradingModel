AI in Algorithmic Trading: A Brief Literature Review

Algorithmic trading, or “algo trading,” uses automated programs to trade in financial markets at high speeds and volumes with minimal human involvement (Chan, 2017). Over the past decade, advances in AI have transformed this field, enhancing predictive accuracy, risk management, and execution strategies. This review highlights how AI is applied to algorithmic trading, focusing on machine learning, deep learning, reinforcement learning, and natural language processing.


1. Traditional Machine Learning Approaches
Early AI-based algorithmic trading used linear regression, support vector machines, and random forests (Hull, 2012). These methods proved effective in predicting short-term price movements and uncovering patterns in historical data (Vanstone & Finnie, 2009). For instance, support vector machines have been used to classify market regimes (e.g., bullish or bearish), while random forests identify top features for forecasting price fluctuations. Despite their contributions, these models often struggle with highly nonlinear market behaviours.


2. Deep Learning Approaches
More recently, deep learning architectures—especially recurrent neural networks (RNNs), long short-term memory (LSTM) networks, and convolutional neural networks (CNNs)—have gained popularity for time-series forecasting and feature extraction (Fischer & Krauss, 2018). LSTMs, in particular, to address the vanishing gradient problem and capture longer-term dependencies in sequential data (Hochreiter & Schmidhuber, 1997). Hybrid models combining deep learning with technical indicators and fundamental analysis also show promise (Zhang, Aggarwal & Zulkernine, 2020). However, overfitting remains a concern, and ensuring robust out-of-sample performance is a key challenge.


3. Reinforcement Learning in Algorithmic Trading
RL helps trading strategies to adapt to the changing market conditions, having an "agent" choose actions—buy, sell, or hold—and receive profits or losses as rewards. Over time, it learns a policy to maximize those rewards (Deng, Bao & Shen, 2017). Moreover, deep RL methods—combining neural networks with algorithms like Deep Q-Networks or Proximal Policy Optimization—additionally consider transaction costs and market impact, holding promise in portfolio management and order execution (Jiang & Liang, 2017). The main drawback with these models, however, lies in their potential computational expensiveness and potential to underperform in drastically changing market conditions.


4. Sentiment Analysis and Natural Language Processing
Financial markets are highly sensitive to news, social media chatter, and macroeconomic indicators (Tetlock, 2007). With a push from AI, NLP techniques can process unstructured textual data, from earnings reports to social media posts, in order to gauge market sentiment (Barbaglia, Croux & Wilms, 2020). Where early approaches to sentiment analysis relied on simple dictionary-based approaches, modern techniques adopt deep learning models like transformers (e.g., BERT) to better capture context-dependent language features.
By adding in the sentiment indicators, algorithmic trading systems can possibly capture the short-term market movement driven by news or social media trends (Peng & Jiang, 2020). The integration has proven especially useful for event-driven trading strategies where speed and accurate interpretation of textual data are critical.

6. Challenges and Future Directions
Despite all the progress, there are several challenges in AI-driven algorithmic trading. Financial time-series data are non-stationary; market regimes change very frequently, making it very difficult for AI models to maintain predictive performance (Lo, 2017). There is always the risk of overfitting, as complex models trained on historical data may not generalize to new market conditions. Another important aspect that is coming under scrutiny is ethical and regulatory aspects in AI systems, such as market manipulation and fairness, as the systems become more autonomous.
Looking ahead, researchers are looking into XAI techniques to make the algorithmic trading models transparent, which may further help market participants and regulators make better, model-driven decisions (Doshi-Velez & Kim, 2017). Under study is also multi-agent RL and more sophisticated transfer learning frameworks to enhance adaptive capabilities under evolving market conditions. Finally, AI solutions integrated with alternative data sources, such as satellite images or geospatial data, are likely to be differentiated and yield alpha-generating insights.

1. Linear Regression
   
Description
Linear Regression is perhaps one of the simplest supervised learning techniques. It models the relationship between a dependent variable, like a stock price, with one or more independent variables—for example, historical prices, trading volume, or selected technical indicators.

How It Applies to Algorithmic Trading
•	Goal: Prediction of future stock prices (or returns) based on a linear combination of historical data and indicators.
•	Process:
    1.	Collect historical stock prices and associated features (technical indicators, market sentiment scores, etc.)
    2.	Split into training and testing sets.
    3.	Fit a linear model to the training data.
    4.	Use the model to predict future prices on the test set.
Strengths
    •	Easy to Use: Simple to understand and implement with widely available libraries (e.g., scikit-learn in Python).
    •	Interpretability: The coefficients show the direction and strength of each feature's impact on the predicted price.
    •	Low Computational Cost: Linear Regression has low overhead, so it's fast to train on large datasets.
Weaknesses
    •	Linearity Assumption: Assumes a linear relationship between features and the target—may not capture complex market dynamics.
    •	Sensitivity to Outliers: Large outliers in stock prices may heavily distort the regression line.
    •	Limited Non-Stationary Handling: Financial time series are usually non-stationary and subject to regime changes; linear models may fail unless  properly adjusted (e.g., using rolling windows or transformations specific to time-series).
Types of Input Data
    •	Historical daily/weekly/monthly stock prices
    •	Trading volume data
    •	Technical indicators (e.g., moving averages, RSI, MACD)
    •	Macroeconomic or fundamental data (optional, if you wish to include broader economic indicators)
Expected Output
    •	Future stock price (point prediction)
    •	Trading signal (if you threshold the predicted return to decide buy/sell/hold)


2. Feedforward Neural Network (Shallow Neural Network)
   
Description
A standard feed-forward neural network, often simply called a "Neural Network," is a collection of interconnected layers of neurons, where each neuron applies a weighted sum and activation function to input data.

How It Applies to Algorithmic Trading
•	Goal: To capture potentially non-linear relationships in stock price movements, which may be missed by simpler methods such as Linear Regression.
•	Process:
    1.	Collect and preprocess historical data similar to regression.
    2.	Design a neural network with one or two hidden layers with an appropriate number of neurons.
    3.	Train the network (commonly using backpropagation) on historical data to minimize prediction error.
    4.	Use the trained network for out-of-sample predictions.
Strengths
    •	Non-linear Learning: Neural networks can learn difficult non-linear mappings from inputs (market data) to outputs (stock prices).
    •	Flexible Architecture: It is relatively easy to extend the architecture by adding layers/neurons if more capacity is needed.
    •	Broad Tooling Support: One of the simplest experiments, with many high-level libraries: TensorFlow, PyTorch, Keras.
Weaknesses
    •	Overfitting: With limited data or too complex networks, the model may memorize training examples rather than learn robust patterns.
    •	Less Interpretable: Neural networks are often considered “black boxes,” making it harder to interpret how inputs influence predictions.
    •	More Computationally Demanding: Training can be slower, especially on large datasets or with deeper networks.
Types of Input Data
    •	Time-windowed historical data (e.g., sliding window of the last 30 days of prices)
    •	Technical indicators as extra input features
    •	Possible textual sentiment scores (e.g., from news or social media), although more complex models may be required
Expected Output
    •	Future price or price change (regression output)
    •	Buy/Sell/Hold signals if you transform the problem into a classification task
    
3. Bayesian Networks

Description
A Bayesian network, or BN, is a probabilistic graphical model, which consists of variables—conventionally known as nodes—and their conditional dependencies, conventionally called edges. Contrary to purely deterministic models, BNs represent uncertainty in terms of dealing with prior knowledge incorporation.

How It Applies to Algorithmic Trading
•	Goal: Model uncertainty in market behavior and update predictions as new evidence (data) becomes available.
•	Process:
    1.	Define your nodes, e.g., stock price level, volatility, and macroeconomic factors.
    2.	Specify conditional probability distributions that relate these variables.
    3.	Use Bayesian inference to update beliefs as new data arrives, for making predictions on future price movements.
Strengths
    •	Uncertainty Quantification: Outputs are probability distributions rather than single point estimates, which can be useful in risk management.
    •	Incorporation of Prior Knowledge: You can encode known relationships between factors (e.g., if macroeconomic indicators strongly influence a sector).
    •	Interpretability: The structure of the network reveals how variables influence each other.
Weaknesses
    •	Complexity of Building the Network: Defining the correct network structure and conditional probabilities can be difficult, especially for high- dimensional data.
    •	Computational Complexity: Inference in large Bayesian networks can become computationally expensive.
    •	Data Requirements: Requires sufficient data to estimate conditional probabilities accurately. Incomplete data or noisy data makes it difficult  to construct the model.
    
Types of Input Data
    •	Discrete or continuous: Stock price movements can be discretized (e.g., up, down, unchanged) or handled as continuous distributions.
    •	Multiple sources of risk/market data: Macroeconomic indicators, industry news, fundamental data, etc.
Expected Output
    •	Probability distribution over potential future price movements (e.g., the probability that the stock will move up by X%, down by Y%, or stay within a small range).
    •	Probabilistic risk assessments, supporting more nuanced decision-making in trade execution.

```python
# A. Setup and Data Collection
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

###############################################################################
# 1. Data Retrieval Function
###############################################################################
def get_data(ticker: str, start_date: str, end_date: str) -> pd.DataFrame:
    """
    Fetches historical stock data from Yahoo Finance using the yfinance library.

    :param ticker: Stock ticker symbol, e.g., 'AAPL'.
    :param start_date: Start date for the data in 'YYYY-MM-DD' format.
    :param end_date: End date for the data in 'YYYY-MM-DD' format.
    :return: A pandas DataFrame containing the downloaded data.
    """
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

###############################################################################
# 2. Data Preprocessing Function
###############################################################################
def preprocess_data(data: pd.DataFrame) -> pd.DataFrame:
    """
    Cleans and prepares the data for feature engineering:
    - Removes any rows with NaN values
    - Resets the index to have a 'Date' column in the DataFrame

    :param data: Raw pandas DataFrame from Yahoo Finance.
    :return: Cleaned pandas DataFrame with a reset index.
    """
    data.dropna(inplace=True)
    data.reset_index(inplace=True)  # Move 'Date' from index to a column
    return data

###############################################################################
# 3. Feature Engineering Function
###############################################################################
def create_features(data: pd.DataFrame) -> pd.DataFrame:
    """
    Creates features for the regression model, including:
    - 'PrevClose': the previous day's closing price
    - 'MA_5': 5-day Moving Average
    - 'MA_10': 10-day Moving Average

    :param data: Clean pandas DataFrame.
    :return: DataFrame with new feature columns.
    """
    # Previous day close
    data['PrevClose'] = data['Close'].shift(1)

    # Moving averages
    data['MA_5'] = data['Close'].rolling(window=5).mean()
    data['MA_10'] = data['Close'].rolling(window=10).mean()

    # Drop rows that have NaN due to shifting/rolling
    data.dropna(inplace=True)

    return data

###############################################################################
# 4. Train/Test Split Function
###############################################################################
def split_data(data: pd.DataFrame, split_ratio: float = 0.8):
    """
    Splits the data into train and test sets while respecting time order.

    :param data: DataFrame containing features and target.
    :param split_ratio: Fraction of the data to use for training (default = 0.8).
    :return: X_train, y_train, X_test, y_test, train_df, test_df
    """
    split_index = int(len(data) * split_ratio)
    train_df = data.iloc[:split_index]
    test_df = data.iloc[split_index:]
    
    # Here we include multiple features in our model
    X_train = train_df[['PrevClose', 'MA_5', 'MA_10']]
    y_train = train_df['Close']
    X_test = test_df[['PrevClose', 'MA_5', 'MA_10']]
    y_test = test_df['Close']
    
    return X_train, y_train, X_test, y_test, train_df, test_df

###############################################################################
# 5. Model Training Function
###############################################################################
def train_model(X_train, y_train) -> LinearRegression:
    """
    Trains a Linear Regression model.

    :param X_train: Feature matrix for training.
    :param y_train: Target vector for training.
    :return: Trained LinearRegression model.
    """
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model

###############################################################################
# 6. Prediction Function
###############################################################################
def make_predictions(model, X_test) -> np.ndarray:
    """
    Generates predictions using the trained model.

    :param model: Trained model (LinearRegression).
    :param X_test: Feature matrix for testing.
    :return: Predicted values as a numpy array.
    """
    return model.predict(X_test)

###############################################################################
# 7. Evaluation Function
###############################################################################
def evaluate_model(y_test, y_pred) -> None:
    """
    Prints evaluation metrics for the model.

    :param y_test: Actual target values.
    :param y_pred: Model predictions.
    """
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f'MSE: {mse:.2f}')
    print(f'MAE: {mae:.2f}')

###############################################################################
# 8. Visualization Function
###############################################################################
def visualize_results(test_df: pd.DataFrame, y_test: pd.Series, y_pred: np.ndarray, 
                      ticker: str) -> None:
    """
    Plots the actual vs. predicted stock prices for the test set.

    :param test_df: DataFrame corresponding to the test set (for the Date axis).
    :param y_test: Actual values of the test set.
    :param y_pred: Predicted values for the test set.
    :param ticker: The stock ticker (used for plot title).
    """
    plt.figure(figsize=(10,5))
    plt.plot(test_df['Date'], y_test, label='Actual Prices')
    plt.plot(test_df['Date'], y_pred, label='Predicted Prices')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(f'{ticker} - Linear Regression Prediction')
    plt.legend()
    plt.show()

###############################################################################
# Main Function
###############################################################################
def main():
    ticker = 'AAPL'  # Example: Apple stock
    start_date = '2023-01-01'
    end_date = '2025-01-06' 

    # A) Data Collection
    data = get_data(ticker, start_date, end_date)

    # B) Data Preprocessing
    data = preprocess_data(data)

    # C) Feature Engineering (now includes MA_5 and MA_10)
    data = create_features(data)

    # D) Train/Test Split
    X_train, y_train, X_test, y_test, train_df, test_df = split_data(data)

    # E) Model Training
    model = train_model(X_train, y_train)

    # F) Make Predictions
    y_pred = make_predictions(model, X_test)

    # G) Evaluate Model
    evaluate_model(y_test, y_pred)

    # H) Visualization
    visualize_results(test_df, y_test, y_pred, ticker)

# If running as a script or in a Jupyter Notebook, call main().
if __name__ == "__main__":
    main()
```

**D.1 Describe the Appropriate Testing Methods**

Train/Test Split with Time-Series Consideration

Because we are dealing with time-series data, we performed a chronological split (e.g., the first 80% of the data for training, the last 20% for testing). This ensures the model is trained on past data and then evaluated on future data, mimicking real-world prediction scenarios.
Alternative: We could use TimeSeriesSplit from sklearn.model_selection for more robust time-series cross-validation. However, the basic 80/20 split still respects the time order, preventing data leakage from the future to the past.
Feature Validation

We tested that our features (e.g., moving averages, previous day’s close) are aligned correctly with the target (close price). For instance, the Close at time t should not be used to predict itself at time t, which would cause data leakage.
Baseline Comparison

A key testing method involves comparing our model against a naive baseline:
For example, “Predict tomorrow’s price = today’s price.”
If our Mean Absolute Error (MAE) or Mean Squared Error (MSE) is significantly lower than this baseline, we know our model is capturing useful patterns.
Residual Analysis

We inspect the residuals (the difference between actual and predicted values) to see if there are patterns we missed.
A random scatter of residuals around zero suggests the model is not systematically biased.
Performance Metrics

We use MAE and MSE as our primary metrics. MAE gives an easily interpretable measure in the same units as the stock price (dollars), while MSE penalizes larger errors more heavily.
Why these methods are appropriate: In finance, ensuring the model’s predictions use only past information is critical. Train/test splitting by date and using robust metrics that measure forecast accuracy are standard practices in time-series forecasting.
